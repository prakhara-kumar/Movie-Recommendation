
Importing the dependencies

```python
import pandas as pd
import numpy as np
```

```python
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre-Processing

```python
mov = pd.read_csv("movies.csv")
```

```python
# Number of rows and columns
mov.shape
```

```python
# Wanted to see the column names
list(mov.columns)
```

```python
mov.head(5)
```

```python
# Selecting relevant features : Feature Selection

features = ['genres', 'keywords', 'tagline', 'overview', 'cast', 'director']
print(features)
```

```python
# Checking for count of null values in the features that we have selected

for i in features : 
    print(i,mov[i].isnull().sum())
```

```python
# Filling null values 

for i in features : 
    mov[i] = mov[i].fillna('')
```

```python
# Combining all the features selected

combine = mov["genres"]+' '+mov["keywords"]+' '+mov["tagline"]+' '+mov["overview"]+' '+mov['cast']+' '+mov["director"]
```

```python
print(combine)
```

```python
# Converting data to vectors

vectors = TfidfVectorizer()
```

```python
# Store all the data in numerical values
veccombine = vectors.fit_transform(combine)
```

```python
print(veccombine)
```

Cosine similarity : Similarity scores

```python
similarity = cosine_similarity(veccombine)
```

```python
print(similarity)
```

```python
# Take one movie and compare it with other movies --> gives us similarity score
similarity.shape
```

```python
# Getting the movie name

movie_name = input("Enter your favourite movie name : ")
```

```python
# Creating a list with all the movies given in the dataset

list_of_movies = mov['title'].tolist()
print(len(list_of_movies))
```

```python
# Finding close match for movie name given by the user 

close_match = difflib.get_close_matches(movie_name, list_of_movies)
print(close_match)
```

```python
match = close_match[0]
print(match)
```

```python
# Finding the index of movie

index_of_matched = mov [mov.title == match] ['index'].values[0]
print(index_of_matched)
```

```python
# Getting alist of similar movies
# (0, 0.050174693392667453) --> means that first value is the index, second value is the similarity score of movie compared to "match"

sim_score = list(enumerate(similarity[index_of_matched]))
print(sim_score)
```

```python
len(sim_score)
```

```python
# Sorting the movies based on similarity scores(desc order)
# lambda is used : x --> sim_score, x[1] -->scores of movie

sorted_sim_movie_score = sorted(sim_score, key = lambda x:x[1], reverse = True)
print(sorted_sim_movie_score)
```

```python
# print the names of similar to the match based on index

print('Movie Suggestion for you : \n')
i = 1

for movie in sorted_sim_movie_score :
    indexx = movie[0]
    
    title_from_index = mov[mov.index == indexx]['title'].values[0]
    
    if (i < 21) :
        print(i, ".", title_from_index)
        i += 1
```
